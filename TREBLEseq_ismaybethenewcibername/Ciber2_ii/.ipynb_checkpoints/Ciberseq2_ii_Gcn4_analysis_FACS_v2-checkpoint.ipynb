```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import csv
import scipy
import string
from scipy.stats import spearmanr
```

```python
# read in all_experiments_controlled_averaged.csv

ciber240 = pd.read_csv('t240ratios.csv', index_col=0)
ciber240['average240'] = (ciber240['AD_3_240']+ ciber240['AD_4_240'])/2
ciber240
```

```python
# read in BYS4 data csv
facs = pd.read_csv('Gcn4_orthologs_TileDF_forTeaching_v4.csv', usecols=['ArrayDNA', 'TotalReads_BioRepA_BYS4','TotalReads_BioRepB_BYS4','GFPsortSignal_BioRepA', 'GFPsortSignal_BioRepB']).fillna(0)
facs['GFP_weightedav'] = ((facs['GFPsortSignal_BioRepA']*facs['TotalReads_BioRepA_BYS4'])+(facs['GFPsortSignal_BioRepB']*facs['TotalReads_BioRepB_BYS4']))/(facs['TotalReads_BioRepA_BYS4']+facs['TotalReads_BioRepB_BYS4'])
facs = facs[['ArrayDNA', 'GFP_weightedav']]
facs = facs.dropna()
facs = facs[facs.GFP_weightedav != 0]
facs
```

```python
intersection240 = pd.merge(ciber240, facs,
                       left_on = 'Tile', right_on = 'ArrayDNA',
                       how= 'inner')
intersection240['GFP_weightedav'] += 10
```

```python
p = sns.jointplot(intersection240, x='average240', y='GFP_weightedav', marker='.')
p.ax_joint.set_xscale('log')
p.ax_joint.set_yscale('log')
plt.savefig('t240_jointplot.jpg', format='jpg')
```

```python
intersection240['average240'].corr(intersection240['GFP_weightedav'])
```

```python
spearmanr(intersection240['average240'], intersection240['GFP_weightedav'])
```
