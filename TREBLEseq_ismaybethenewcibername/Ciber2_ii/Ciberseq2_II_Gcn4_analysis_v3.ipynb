## Individual transcript dynamics

```python
def extract_tile(putative_tile):
    # Remove the surrounding square brackets and quotes
    clean_str = putative_tile.strip("[]").replace("'", "")
    # Split by hyphen and take the first part
    tile_str = clean_str.split('-')[0]
    return tile_str
```

```python
# concatenate RPTR dataframes of each sample
Rsample2 = []
Rsample3 = []
Rsample4 = []
for key in RPTRexperiments_dict.keys():
    if key.startswith('RPTR_2') :
#         print(key)
        RPTRexperiments_dict[key]['Tile'] = RPTRexperiments_dict[key]['PutativeTileADBC'].apply(extract_tile)
        Rsample2.append(RPTRexperiments_dict[key])
    if key.startswith('RPTR_3') :
#         print(key)
        RPTRexperiments_dict[key]['Tile'] = RPTRexperiments_dict[key]['PutativeTileADBC'].apply(extract_tile)     
        Rsample3.append(RPTRexperiments_dict[key])
    if key.startswith('RPTR_4') :
#         print(key)
        RPTRexperiments_dict[key]['Tile'] = RPTRexperiments_dict[key]['PutativeTileADBC'].apply(extract_tile)
        Rsample4.append(RPTRexperiments_dict[key])  
        
Rexp2 = pd.concat(Rsample2, ignore_index=True)
Rexp3 = pd.concat(Rsample3, ignore_index=True)
Rexp4 = pd.concat(Rsample4, ignore_index=True)

Rexp2['R_Normalized'] = Rexp2['BCs']/Rexp2['Allreads']
Rexp3['R_Normalized'] = Rexp3['BCs']/Rexp3['Allreads']
Rexp4['R_Normalized'] = Rexp4['BCs']/Rexp4['Allreads']

Rexp2_facs = pd.merge(Rexp2, facs,left_on = 'Tile', right_on = 'ArrayDNA',)[['index', 'Library', 'R_Normalized', 'is_control']]
Rexp3_facs = pd.merge(Rexp3, facs,left_on = 'Tile', right_on = 'ArrayDNA',)[['index', 'Library', 'R_Normalized', 'is_control']]
Rexp4_facs = pd.merge(Rexp4, facs,left_on = 'Tile', right_on = 'ArrayDNA',)[['index', 'Library', 'R_Normalized', 'is_control']]


Rexp4_facs.head()
```

```python
allRPTRlist = [Rexp2, Rexp3, Rexp4]
allRPTRdf = pd.concat(allRPTRlist, ignore_index=True)
allRPTR_facs = pd.merge(allRPTRdf, facs, left_on = 'Tile', right_on = 'ArrayDNA')
allRPTR_facs.shape
```

## What is the difference between RPTR transcript counts before and after induction?

### How do RPTR reads change over time? Focus on controls

```python
#transform data so that columns are RPTR counts at each time point
Rexp4_collapsed= Rexp4_facs.pivot_table(index=['index', 'is_control'], columns='Library', values='R_Normalized').reset_index().fillna(1e-10)
Rexp4_collapsed = Rexp4_collapsed.set_index(['index', 'is_control'])
#put columns in time order:
Rexp4_collapsed = Rexp4_collapsed[['RPTR_4_0', 'RPTR_4_5', 'RPTR_4_10', 'RPTR_4_15', 'RPTR_4_30', 'RPTR_4_180', 'RPTR_4_240']]

Rexp4_collapsed.head()
```

```python
Rexp4_collapsed.shape
```

```python
intersection4_gpc = pd.DataFrame()
intersection4_gpc['005'] = ((Rexp4_collapsed['RPTR_4_5']-Rexp4_collapsed['RPTR_4_0'])/(Rexp4_collapsed['RPTR_4_0'])*100)
intersection4_gpc['010'] = ((Rexp4_collapsed['RPTR_4_10']-Rexp4_collapsed['RPTR_4_0'])/(Rexp4_collapsed['RPTR_4_0'])*100)
intersection4_gpc['015'] = ((Rexp4_collapsed['RPTR_4_15']-Rexp4_collapsed['RPTR_4_0'])/(Rexp4_collapsed['RPTR_4_0'])*100)
intersection4_gpc['030'] = ((Rexp4_collapsed['RPTR_4_30']-Rexp4_collapsed['RPTR_4_0'])/(Rexp4_collapsed['RPTR_4_0'])*100)
intersection4_gpc['180'] = ((Rexp4_collapsed['RPTR_4_180']-Rexp4_collapsed['RPTR_4_0'])/(Rexp4_collapsed['RPTR_4_0'])*100)
intersection4_gpc['240'] = ((Rexp4_collapsed['RPTR_4_240']-Rexp4_collapsed['RPTR_4_0'])/(Rexp4_collapsed['RPTR_4_0'])*100)

intersection4_gpc.head()
```

```python
intersection4_gpc_melt = intersection4_gpc.reset_index()
intersection4_gpc_melt = intersection4_gpc_melt.melt(id_vars=['index','is_control'], var_name='Time', value_name='Pct_change')
intersection4_gpc_melt.head()
```

```python
# Plot the data
# plt.figure(figsize=(10, 6))
sns.lineplot(data=intersection4_gpc_melt, 
             x='Time', y='Pct_change', 
             hue='index', marker='o', palette='mako', 
             legend=False
            )

# Set y-axis to log scale
plt.yscale('symlog')

plt.title('Molecule Abundance after induction')
plt.xlabel('Time Points')
plt.ylabel('Pct_Change (log)')
plt.show()
```
