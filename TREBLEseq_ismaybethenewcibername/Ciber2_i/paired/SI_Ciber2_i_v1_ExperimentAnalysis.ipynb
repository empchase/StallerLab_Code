```python
csv_directory = os.path.join(os.getcwd(), 'ExperimentData_20240423')
experiments_dict = {}

# Iterate over files in the directory
for filename in os.listdir(csv_directory):
#     print(filename)
    # Check if the file is a regular file
    if os.path.isfile(os.path.join(csv_directory, filename)):
        # Assuming all files are CSV, adjust the condition as needed
        if filename.endswith('.csv'):
            # Read the file into a DataFrame
            df = pd.read_csv(os.path.join(csv_directory, filename), usecols=['index_x', 'BCs', 'Library_x', 'index_y', 'Tile-AD','Ratio'])
            # Get the value in column x of row 1
            key_value = df.iloc[0]['Library_x']
            # Store the DataFrame in the dictionary with the key as the value in column x of row 1
            experiments_dict[key_value] = df
            print(key_value)

# Now 'dfs' contains a dictionary where keys are the values in column x of row 1, and values are the corresponding DataFrames
```

```python
tmp = pd.merge(experiments_dict['Staller_RPTR_3-0-2_MVS_0078_I1'], 
               experiments_dict['Staller_RPTR_3-5-2_MVS_0079_I1'], 
              on = 'index_x',
              suffixes=('_df_0', '_df_5'),
              how = 'inner')
p = sns.regplot(tmp, x='Ratio_df_0', 
                     y='Ratio_df_5')
p.set_title('RPTR_3-5 vs RPTR_3-0')
p.set_xlabel('RPTR_3-0')
p.set_ylabel('RPTR_3-5')

slope, intercept, r, p, sterr = scipy.stats.linregress(x=p.get_lines()[0].get_xdata(),
                                                       y=p.get_lines()[0].get_ydata())
print(slope, intercept)
```

```python
correlation_5 = tmp['Ratio_df_0'].corr(tmp['Ratio_df_5'])
correlation_5
```
